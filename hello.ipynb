```python
import os
import pickle
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import undetected_chromedriver as uc

# File to store cookies
COOKIES_FILE = "google_messages_cookies.pkl"

# URL of Google Messages
GOOGLE_MESSAGES_URL = "https://messages.google.com/web/"

def save_credentials():
    """
    Launches the browser for the first-time login and saves cookies.
    """
    # Chrome options for visible login
    options = Options()
    options.add_argument("--start-maximized")
    driver = uc.Chrome(options=options)

    try:
        # Navigate to Google Messages
        driver.get(GOOGLE_MESSAGES_URL)
        print("Please log in manually...")

        # Wait for the user to log in
        input("Press Enter after completing the login process...")

        # Save cookies after login
        cookies = driver.get_cookies()
        with open(COOKIES_FILE, "wb") as f:
            pickle.dump(cookies, f)
        print("Cookies saved successfully!")

    finally:
        driver.quit()
def send_message(recipient, message):
    """
    Uses saved cookies to send a message through Google Messages in headless mode.
    """
    if not os.path.exists(COOKIES_FILE):
        print("Error: Credentials not found. Run save_credentials() first.")
        return

    # Chrome options for headless mode
    options = Options()
    options.add_argument("--headless=new")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    driver = uc.Chrome(options=options)

    try:
        # Navigate to Google Messages
        driver.get(GOOGLE_MESSAGES_URL)

        # Load cookies
        with open(COOKIES_FILE, "rb") as f:
            cookies = pickle.load(f)
            for cookie in cookies:
                driver.add_cookie(cookie)

        # Refresh the page to load the session
        driver.refresh()
        time.sleep(5)  # Wait for the session to load

        # Click the "Start Chat" button
        start_chat_button = driver.find_element(By.XPATH, '//button[contains(text(), "Start chat")]')
        start_chat_button.click()
        time.sleep(2)

        # Enter the recipient
        recipient_field = driver.find_element(By.XPATH, '//input[@placeholder="Type a name, phone number or email address"]')
        recipient_field.send_keys(recipient)
        recipient_field.send_keys(Keys.ENTER)
        time.sleep(2)

        # Enter the message in the textarea
        message_field = driver.find_element(By.XPATH, '//textarea[@placeholder="Text message"]')
        message_field.send_keys(message)
        time.sleep(1)

        # Locate the send button
        send_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//button[@data-e2e-send-text-button]'))
        )

        # Click the send button
        send_button.click()
        print("Message sent successfully!")

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        driver.quit()

# Example Usage
if __name__ == "__main__":
    # Step 1: Save credentials (Run this only once)
    # save_credentials()

    # Step 2: Send message (Use saved credentials)
    send_message("8438288180", "Hello")
```
